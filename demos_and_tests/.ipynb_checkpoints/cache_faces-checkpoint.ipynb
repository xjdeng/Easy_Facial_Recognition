```python
#Initial imports
from Easy_Image import detect, test
import Easy_Image
import time
```

```python
# Load all images
test_images = detect.load_image_dir("tests/")
print(len(test_images))
```

```python
# Clear faces stored in EXIF (just in case)
test_images.remove_faces()
```

```python
# Show an image in the test_images list:
img5 = test_images[5]
img5.plot()
```

```python
# Time how long it takes to detect faces in all of the images
t0 = time.time()
faces = test_images.detect_faces()
t1 = time.time()
print(str(t1 - t0) + " seconds to detect all faces from scratch")
```

```python
# Show an example face
face5 = faces[5]
face5.plot()
```

```python
# Show an example face using a green bounding box

# First, get the original image that this face came from:
orig_image = face5.parent_image

# Then show this image
orig_image.plot()
```

```python
# Now draw the green bounding box

image_boxes = orig_image.draw_faces()
image_boxes.plot()
```

```python
# Just to make sure, this isn't in place, let's look at the original img again
orig_image.plot()
```
